```python
import json
```

```python
with open("Philippe-II.json", "r+") as jsonFile:
	old_js = json.load(jsonFile)
			
```

```python
def printjs(s):
    print(json.dumps(s, indent=4, sort_keys=True))
```

```python
def clean(my_dict):
    try:
        print(list(my_dict.keys()))
        for k in list(my_dict.keys()):
            if my_dict[k] in ['',{},None]:
                del my_dict[k]
            else:
                my_dict[k]=clean(my_dict[k])
        return my_dict
    except AttributeError:
        print('no keys')
        return my_dict
        
```

```python
clean(old_js)
```

```python
for n in old_js:
    if n['id']!='':
        n['name']=n['firstName']+' '+n['lastName']
        new_js[n['name']]=n
```

```python
new_new_js=clean(new_js)
```

```python
with open("new_js.json", "w+") as jsonFile:
		jsonFile.seek(0) 
		json.dump(new_new_js, jsonFile, sort_keys=True, indent=4, separators=(',', ': '))
		jsonFile.truncate()
```
